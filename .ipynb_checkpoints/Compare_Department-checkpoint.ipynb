```python
import os
import pandas as pd
from loginCirclek import get_session_token
from ExportFile import export, download_exported_file

# Base path
base_path = r'C:\Users\SG0xxx-Tablet\Documents\SM\TOOL BCS\Order\order-backend'
ENDPOINT = "classifiedSaleReport"
SESSION_ID = get_session_token()

folder_path = os.path.join(base_path, 'Data', 'Department')
os.makedirs(folder_path, exist_ok=True)

def clean_excel(file_path, period, startDate, endDate):
    """Đọc file, loại bỏ 3 dòng đầu, xóa cột thừa và thêm period/startDate/endDate."""
    df = pd.read_excel(file_path, header=None, engine="openpyxl")

    # Bỏ 3 dòng đầu
    df = df.iloc[2:].reset_index(drop=True)

    # Gán header từ dòng đầu tiên sau khi bỏ 3 dòng
    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)

    # Xóa các cột không cần
    cols_to_drop = [
        "Store Name", 
        "Top Department", 
        "Area Manager ID", 
        "Area Manager Name", 
        "Store No.", 
        "NO"
    ]
    df = df.drop(columns=[c for c in cols_to_drop if c in df.columns])

    # Thêm thông tin kỳ
    df["period"] = period
    df["startDate"] = startDate
    df["endDate"] = endDate

    return df


# Ngày this
startDate, endDate = "20250901", "20250918"
exp_key_this = export(startDate, endDate, SESSION_ID, ENDPOINT)
download_exported_file(exp_key_this, SESSION_ID, filename="this", folder_path=folder_path)
file_this = os.path.join(folder_path, "this.xlsx")
df_this = clean_excel(file_this, "this", startDate, endDate)

# Ngày last
lastStart, lastEnd = "20240801", "20240818"
exp_key_last = export(lastStart, lastEnd, SESSION_ID, ENDPOINT)
download_exported_file(exp_key_last, SESSION_ID, filename="last", folder_path=folder_path)
file_last = os.path.join(folder_path, "last.xlsx")
df_last = clean_excel(file_last, "last", lastStart, lastEnd)

# Gộp 2 file
df_all = pd.concat([df_this, df_last], ignore_index=True)

# Bỏ cột startDate, endDate (vì đã có Sales Date)
df_all = df_all.drop(columns=["startDate", "endDate"], errors="ignore")

# Xuất file gộp
output_file = os.path.join(folder_path, "merged_department.xlsx")
df_all.to_excel(output_file, index=False)

print(f"File đã được gộp: {output_file}")
print(df_all.head())
```
